```python
import pandas as pd
import glob
import os

# 엑셀 파일들이 저장된 폴더 경로
folder_path = 'data/'

# 폴더 내 모든 엑셀 파일 경로 가져오기
excel_files = glob.glob(folder_path + '*.xls')

# 모든 엑셀 파일을 읽어서 리스트에 저장
dfs = []
filtered = []

for file in excel_files:
    try:
        df = pd.read_excel(file, skiprows=3)
        print(f"파일 로드 완료: {os.path.basename(file)}")
        dfs.append(df)
        n = len(dfs)-1
        # 첫 번째 DataFrame 사용
        if isinstance(dfs, list) and n > 0:
            filtered_df = dfs[n][dfs[n]['시구'].astype(str) == '전체(시도)']
            filtered.append(filtered_df)

    except Exception as e:
        print(f"파일 로드 실패: {file}, 오류: {e}")
        
# 필터링된 데이터프레임들을 변수명 제외하고 합치기
if filtered:
    # 첫 번째 데이터프레임의 열 이름을 기준으로 설정
    base_columns = filtered[0].columns
    aligned_dfs = []
    
    for df in filtered:
        df_copy = df.copy()
        # 열 개수가 같은 경우에만 열 이름 변경
        if len(df_copy.columns) == len(base_columns):
            df_copy.columns = base_columns
        aligned_dfs.append(df_copy)
    
    # 행 방향으로 합치기
    final_result = pd.concat(aligned_dfs, ignore_index=True)
```

```python
final_result.describe()
```

```python
final_result['고객호수(호)'] = final_result['고객호수(호)'].str.replace(',', '').astype(float)
final_result['평균판매단가(원/kWh)'] = pd.to_numeric(final_result['평균판매단가(원/kWh)'], errors='coerce')
```

```python
print(final_result.dtypes)  # 각 컬럼의 데이터 타입 확인
```

```python
result = final_result.groupby(['년월', '시구']).agg({
    '고객호수(호)': 'sum', 
    '평균판매단가(원/kWh)': 'mean'
}).reset_index()

result['년월'] = result['년월'].astype(str)
result['년월'] = pd.to_datetime(result['년월'], format='%Y%m')
```

```python
import numpy as np

CPI = pd.read_csv('CPI.csv', encoding='cp949')
PPI = pd.read_csv('PPI.csv', encoding='cp949')
```

```python
CPI = CPI.rename(columns={"__": "년월", "_뎅": "CPI"})
PPI = PPI.rename(columns={"시점": '년월', '총지수 (2020=100)': 'PPI'})
```

```python
CPI['년월'] = CPI['년월'].apply(lambda x: f'{float(x):.2f}').astype(str)
CPI['년월'] = CPI['년월'].str.replace('.', '')
CPI['년월'] = pd.to_datetime(CPI['년월'], format='%Y%m')

PPI['년월'] = PPI['년월'].apply(lambda x: f'{float(x):.2f}').astype(str)
PPI['년월'] = PPI['년월'].str.replace('.', '')
PPI['년월'] = pd.to_datetime(PPI['년월'], format='%Y%m')
```

```python
merged_result = pd.merge(result, CPI, on = '년월', how='left').merge(PPI, on = '년월', how = 'left')
```

```python
merged_result['실질판매단가_CPI'] = 100 / merged_result['CPI'] * merged_result['평균판매단가(원/kWh)']
merged_result['실질판매단가_PPI'] = 100 / merged_result['PPI'] * merged_result['평균판매단가(원/kWh)']
```

```python
merged_result
```

```python
import matplotlib.pyplot as plt

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(merged_result['년월'],merged_result['실질판매단가_CPI'], marker='o', linestyle='-', linewidth=2, markersize=6)
plt.plot(merged_result['년월'],merged_result['실질판매단가_PPI'], marker='s', linestyle='-', linewidth=2, markersize=6)
plt.plot(merged_result['년월'],merged_result['평균판매단가(원/kWh)'], marker='^', linestyle='--', linewidth=2, markersize=6)
plt.title('Trend of the electricity price', fontsize=16, fontweight='bold')
plt.xlabel('time', fontsize=12)
plt.ylabel('real Price', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# 그래프 표시
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

plt.figure(figsize=(12, 6))

# 원본 데이터 플롯 (선만)
plt.plot(merged_result['년월'], merged_result['실질판매단가_CPI'], 
         linestyle='-', linewidth=2, label='실질판매단가_CPI', alpha=0.7)
plt.plot(merged_result['년월'], merged_result['실질판매단가_PPI'], 
         linestyle='-', linewidth=2, label='실질판매단가_PPI', alpha=0.7)

# 추세선 추가 (예: 실질판매단가_CPI)
x = np.arange(len(merged_result)).reshape(-1, 1)
y = merged_result['실질판매단가_CPI'].values
reg = LinearRegression().fit(x, y)
trend_line = reg.predict(x)

plt.plot(merged_result['년월'], trend_line, 
         linestyle=':', linewidth=3, color='red', label='CPI 추세선')

plt.title('Trend of the electricity price', fontsize=16, fontweight='bold')
plt.xlabel('time', fontsize=12)
plt.ylabel('real Price', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
